##まずはトランプのカードを作る
<div style="display:lock;">
ブラックジャックはトランプを使ったカードゲームなので、何をおいてもまずトランプが必要です。</div>

<br>
トランプはマークが♥︎♦︎♣️♠︎の４種類、それぞれに１〜１３の数字があります。  
  
今回ジョーカーは使わないので、計５２枚のカードがあればいいわけです。  
<br>
では実際にパソコン上にトランプのカードを作っていきましょう。
  
ここでバカ正直に５２個の<font color="pink">オブジェクト</font>を作ってもいいのですが、  

面倒なので白紙の<font color="skyblue">【トランプカード】</font>という<font color="pink">オブジェクト</font>を作ります。  

プログラミングでは、少しずつ違うものを５２個用意するよりも、  
  
同じものを５２個用意する方が簡単に書けるからです。  
<br>
この白紙の<font color="skyblue">【トランプカード】</font>を５２枚コピーして、後から  
<font color="lime">《マーク》</font>と<font color="lime">《数》</font>の入った<font color="lime">《表面》</font>を印刷するイメージです。  
<br>
裏面は全てのカードに共通の絵柄なので、今回は考えません。 

```
#トランプカードのマークをリストアップしておきます。
トランプのマーク = ['♥︎', '♦︎', '♣️', '♠︎']

# トランプカードというオブジェクトを作ります。
class トランプカード:
  
  # 新品のトランプを上から順に０〜５１まで番号を振ります。
  def __init__(self, カード番号): #カードの番号に応じて表面を印刷します
    self.表面 = {
        'マーク': トランプのマーク[カード番号//13],
        '数字': カード番号%13+1
    }
    
  # このカードが何のカードか表示する機能を追加します。
  def カード情報(self):
    print(self.表面['マーク'], self.表面['数字'])
```

##トランプカードのテスト


```
#ハートのエース（０番）から始まって、最後はスペードのキング（５１番）です。
#試しにトランプカードを適当に生成します。

ハートのエース = トランプカード(0)
スペードのキング = トランプカード(51)

ハートのエース.カード情報()
スペードのキング.カード情報()
```

##次に山札を作る
ブラックジャックでは５２枚組のトランプを４セット使って山札を作ります。  
  
この<font color="skyblue">【山札】</font>の<font color="pink">オブジェクト</font>もパソコン上に作ってしまいましょう。
  
<font color="skyblue">【山札】</font>は計２０８枚の<font color="lime">《カードの束》</font>を持っています。  
<br>
カードが順番に並んでいたら次に何のカードが出るか分かってしまうので、  
  
ついでに<font color="violet">『シャッフル』</font>機能をつけましょう。  
  
さらに後々のために、山札が今何枚残っているかすぐ分かるように、
  
<font color="violet">『残り枚数』</font>を調べる機能もつけます。


```
import random

#トランプカードを５２枚×４セット使って山札を作る
class 山札:
  def __init__(self, count):
    self.カードの束 = []
    for セット in range(count):
      for カード番号 in range(52):
        self.カードの束.append(トランプカード(カード番号))
        
  #シャッフル機能をつける
  def シャッフル(self):
    シャッフル後 = []
    while len(self.カードの束) > 0:
      #ランダムにカードを抜き出して新しい束にする
      カード = self.カードの束.pop(random.randrange(0, len(self.カードの束)))
      シャッフル後.append(カード)
    self.カードの束 = シャッフル後
    
  #残り枚数をカウント
  def 残り枚数(self):
    return len(self.カードの束)
  
```

##山札のテスト

```
山 = 山札(4)
for カード in 山.カードの束:
  カード.カード情報()
```

```
山.シャッフル()
for カード in 山.カードの束:
  カード.カード情報()
```

## いよいよルールの確認


```
def ブラックジャック式合計(手札):
  合計 = 0
  エースの枚数 = 0
  for カード in 手札:
    if カード.表面['数字'] == 1:
      エースの枚数 += 1
    elif カード.表面['数字'] <= 10:
      合計 += カード.表面['数字']
    else: 
      合計 += 10
  if エースの枚数 > 0:
    while エースの枚数 > 0:
      if 合計+エースの枚数-1 <= 10:
        合計 += 11
      else:
        合計 += 1
      エースの枚数 -= 1
  return 合計

def 勝敗判定(プレイヤー, ディーラー):
  if プレイヤー.決断 == 'サレンダー':
    print('プレイヤーがゲームを降りました', end="\n\n")
    return 0.5
  if プレイヤー.合計値 > 21: # プレイヤーのバーストなのでプレイヤーの負けです
    print('プレイヤーのバースト負け', end="\n\n")
    return 0
  elif ディーラー.合計値 > 21: # ディーラーのバーストなのでプレイヤーの勝ちです
    print('ディーラーがバーストしたのでプレイヤーの勝ち', end="\n\n")
    return 2
  elif プレイヤー.合計値 < ディーラー.合計値: # ディーラーの方が２１に近いのでプレイヤーの負けです
    print('プレイヤーの負け', end="\n\n")
    return 0
  elif プレイヤー.合計値 > ディーラー.合計値: # プレイヤーの方が２１に近いのでプレイヤーの勝ちです
    print('プレイヤーの勝ち', end="\n\n")
    return 2
  else:
    if ディーラー.合計値 == 21:
      if ディーラー.枚数()== 2: # ディーラーがブラックジャックなのでプレイヤーの負けです
        print('ディーラーがブラックジャックなのでプレイヤーの負け', end="\n\n")
        return 0
      elif プレイヤーの手札.枚数() == 2: # プレイヤーがブラックジャックなのでプレイヤーの勝ちです
        print('プレイヤーがブラックジャックなのでプレイヤーの勝ち', end="\n\n")
        return 2.5
    print('引き分け', end="\n\n") # それ以外は引き分けです
    return 1
   
```

##プレイヤーを作る


```
class プレイヤー:
  def __init__(self, 最初のチップ):
    self.チップ = 最初のチップ
    self.手札 = []
    self.合計値 = 0
    self.決断 = ''
    self.属性 = 'プレイヤー'
    
  def カードを引く(self, 山, 枚数):
    for 数 in range(0, 枚数):
      self.手札.append(山.カードの束.pop(0))
    self.合計値 = ブラックジャック式合計(self.手札)
    
  def 枚数(self):
    return len(self.手札)
  
  def カード表示(self, 全表示 = True):
    print('-----', self.属性, '手札-----')
    print('|', end="")
    if self.属性 == 'プレイヤー' or 全表示:
      for カード in self.手札:
        print(' ', カード.表面['マーク'],カード.表面['数字'] , ' |', end="")
      print('    合計：', self.合計値, end="\n\n")
    elif self.属性 == 'ディーラー':
      print(' ', self.手札[0].表面['マーク'], self.手札[0].表面['数字'] , ' |', end="\n\n")
    else:
      print(' ??? |', end="\n\n")
      
  #手札と合計値をリセット
  def リセット(self):
    self.手札 = []
    self.合計値 = 0
```

##最後にディーラーを作る


```
class ディーラー(プレイヤー):
  def __init__(self, 最初のチップ):
    super().__init__(最初のチップ)
    self.属性 = 'ディーラー'
    
  def 考える(self):
    ##手札の合計が１６以下の場合は強制的にヒット
    if self.合計値 <= 16:
      self.決断 = 'ヒット'
    else:
      self.決断 = 'スタンド'
    
```

##ディーラーのテスト

```
敵キャラ = ディーラー(100)
敵キャラ.カードを引く(山, 1)
敵キャラ.カードを引く(山, 1)
print("*****ディーラーの手札*****")
for 敵カード in 敵キャラ.手札:
  敵カード.カード情報()
敵キャラ.考える()
print("-----ディーラーの判断-----")
print(敵キャラ.決断)
```

## ゲームプレイ


```
トランプの山 = 山札(4)
トランプの山.シャッフル()

あなた = プレイヤー(20)
敵 = ディーラー(10000)

賭け金 = 2
配当金 = 0

#ゲームスタート
#山札が無くなるか、プレイヤーのチップが無くなるまで繰り返します。
while トランプの山.残り枚数() > 0 and あなた.チップ >= 賭け金:
  #互いの手札をリセットします。
  あなた.リセット()
  敵.リセット()
  
  #チップをベットします。
  あなた.チップ -= 賭け金
  print('現在のチップは', あなた.チップ, '枚です')
  print('')
  
  #まずは、プレイヤーとディーラーが２枚ずつカードを引きます。
  あなた.カードを引く(トランプの山, 2)
  敵.カードを引く(トランプの山, 2)
  
  #プレイヤーの手札を見ます。
  あなた.カード表示()

  #ディーラーは１枚だけカードを公開します。
  敵.カード表示(False)
  
  #プレイヤーがブラックジャックなら、プレイヤーは何もする必要はありません。
  続行 = あなた.合計値 != 21
  
  #プレイヤーがヒットを止めるか、バーストするまで以下の作業を繰り返します。
  while 続行:
    #現在の手札を見て、次に打つ手を決めてください。
    #ヒット...カードを引く。
    #スタンド...カードを引くのを止める。
    
    あなた.決断 = input('>')
    print('')
    
    #ヒットを選んだら、カードを引きます。
    if あなた.決断 == 'ヒット':
        あなた.カードを引く(トランプの山, 1)
        あなた.カード表示()
        
        if あなた.合計値 >= 21:
          続行 = False
    else :
      続行 = False
  
  #ディーラーのターンです。
  続行 = True
  while 続行:
    敵.考える()
    
    if 敵.決断 == 'ヒット':
      敵.カードを引く(トランプの山, 1)
      
    elif 敵.決断 == 'スタンド':
      続行 = False
      
  #判定です。    
  敵.カード表示()
  配当金 = int(賭け金*勝敗判定(あなた, 敵))
  あなた.チップ += 配当金
  
  続行 = input("勝負を続けますか？(はい/いいえ)：") == 'はい'
  if 続行 == False:
    break
  print('')
  
#最終的に手持ちのチップが増えれば勝利、減ったら負け
if あなた.チップ > 20:
  print('あなたの勝利')
else:
  print('ディーラーの勝利')
```
